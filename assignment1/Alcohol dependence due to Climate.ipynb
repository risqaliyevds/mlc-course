# Does the high consumption of alcoholic beverages depend on the temperature of the country?

## Hypothesis: Alcohol addiction due to climate

```python
# Necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Load datasets

```python
# Source 1. https://datahub.io/five-thirty-eight/alcohol-consumption

# Dataset shows the average per capita consumption of alcoholic beverages by countries in 2010.

df_drinks = pd.read_csv('drinks.csv')
df_drinks.head()
```

```python
# Source 2. https://www.kaggle.com/code/akshaychavan/average-temperature-per-country-per-year/data

# Average temperatures observed in countries between 2000 and 2013

df_temp = pd.read_csv('average_temprature.csv', index_col= 0)
df_temp.head()
```

```python
# Transpose df_temp

df_temp = df_temp.T
df_temp.head()
```

```python
# We only need data for 2010

df_temp_2010 = df_temp[[2010]]
df_temp_2010.head()
```

```python
# Changing the name of the states to General view

df_temp_2010.index = [x.title() for x in df_temp_2010.index]
df_drinks.country = [y.title() for y in df_drinks.country]
```

```python
df_temp_2010.columns.name = ""
df_temp_2010['country'] = df_temp_2010.index[:]
df_temp_2010.index = np.arange(1, 245)
df_temp_2010.head()
```

```python
df_drinks.head()
```

```python
# Combine two datasets (Inner Join)

data = pd.merge(df_drinks, df_temp_2010, how = 'outer', left_on= 'country', right_on='country')
data.head(10)
```

## Data cleaning correcting some values

```python
# Discard rows with all values 0
data.columns = ['Country', 'Beer_servings', 'Spirit_servings', 'Wine_servings', 'Total_alcohol', 'Average_temp']
data.sort_values(by='Country')

data = data[(data['Beer_servings'] != 0) & (data['Spirit_servings'] != 0) & (data['Wine_servings'] !=0)
     & (data['Wine_servings'] != 0) & (data['Total_alcohol'] != 0)]
```

### Maching unmached values

```python
# 1. I fill NaN values with -1 and filter it because there are unmatched values due to orthographic errors when merging
data = data.fillna(-1)

pd.set_option('display.max_rows', None)

# Here is -1 values from drinks table
data_null = data[(data['Beer_servings'] == -1) | (data['Spirit_servings'] == -1) |(data['Country'] == -1)]
data_null
```

```python
# # Here is -1 values from temperature table
data_temp_null = data[data['Average_temp'] == -1]
data_temp_null
```

```python
# Maching values (In some cases i could use loop for changing 'and' to & but it was only three)

# Russia
data.loc[[141],['Country']] = 'Russia'
data.loc[[141],['Average_temp']] = data[data.index == 246]['Average_temp'][246]

# Bosnia & Herzegovina
data.loc[[21],['Country']] = 'Bosnia & Herzegovina'
data.loc[[21],['Average_temp']] = data[data.index == 203]['Average_temp'][203]

# Cote D'Ivoire
data.loc[[28],['Average_temp']] = data[data.index == 206]['Average_temp'][206]

# Congo
data.loc[[47],['Country']] = 'Congo'
data.loc[[47],['Average_temp']] = data[data.index == 210]['Average_temp'][210]

# Cabo Verde
data.loc[[29],['Average_temp']] = data[data.index == 207]['Average_temp'][207]

# Antigua & Barbuda
data.loc[[5],['Average_temp']] = data[data.index == 198]['Average_temp'][198]

# Sao Tome And Principe
data.loc[[148],['Average_temp']] = data[data.index == 253]['Average_temp'][253]

# Timor Leste
data.loc[[171],['Country']] = 'Timor Leste'
data.loc[[171],['Average_temp']] = data[data.index == 259]['Average_temp'][259]

# Trinidad & Tobago
data.loc[[174],['Average_temp']] = data[data.index == 260]['Average_temp'][260]

# United States
data.loc[[184],['Country']] = 'United States'
data.loc[[184],['Average_temp']] = data[data.index == 263]['Average_temp'][263]
```

```python
# Drop other uncorrected -1 values

data = data[(data['Beer_servings'] != -1) & (data['Average_temp'] != -1)]
data
```

```python
# Reindexing

data.index = data['Country']
del data['Country']
data
```

```python
plt.figure(figsize=(16,6))

plt.xticks(rotation = 45)
sns.barplot(y = data.sort_values('Total_alcohol', ascending=False)['Total_alcohol'][:20],
            x = data.sort_values('Total_alcohol', ascending=False).index[:20],
            data = data,
            palette = 'GnBu_r')


plt.title("Top 20 countries with high alcohol consumption", fontsize = 14)
plt.xlabel("Countries", fontsize = 12)
plt.ylabel("Total Litres of pure Alcohol", fontsize = 12)

plt.show()
```

```python
plt.figure(figsize=(16,6))

plt.xticks(rotation = 45)
sns.barplot(y = data.sort_values('Spirit_servings', ascending=False)['Spirit_servings'][:20],
            x = data.sort_values('Spirit_servings', ascending=False).index[:20],
            data = data,
            palette = 'GnBu_r')


plt.title("Top 20 countries that consume the Spirt", fontsize = 14)
plt.xlabel("Countries", fontsize = 12)
plt.ylabel("Spirt servings", fontsize = 12)

plt.show()
```

###  In the above 2 barplots you can see the top 20 Alcohol and Spirt consuming countries. Considering that most of the countries are located in north and the air temperature there is low, it can be seen that alcohol consumption is high in countries with cold climate.

```python
plt.title("Correlation between temperatur and drinks")
sns.heatmap(data.corr(),
            cmap= 'GnBu_r',
            vmin= -1)
plt.xticks(rotation = 45)

plt.show()
```

```python
# Correlatsion

data.corrwith(data.Average_temp)
```

### We have negative correaltion, that is, it can be concluded that consumption increased as Climit decreased. A strong negative correlation of Wine_servings, Total_alcohol columns with Average_temp column can also be seen. As a conclusion, we can see that the hypothesis is reasonable.
